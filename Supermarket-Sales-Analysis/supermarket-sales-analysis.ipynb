**8. Which Product Line is Purchased In The Highest Quantity?**

```python
df.groupby('Product line')['Quantity'].sum().plot(kind='bar')
plt.title('Total Quantity by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Quantity')
plt.show()
```

**9. Display Daily Sales By Day of The Week**

```python
dw_mapping={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    7:'Sun'
}
```

```python
df['day_of_week']=df['Date'].dt.dayofweek.map(dw_mapping)
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
```

**10. What Will Be The Highest Months For Sales**

```python
month_mapping={
    1:'Jan',
    2:'Feb',
    3:'March'
}
```

```python
df['month']=df['Date'].dt.month.map(month_mapping)
```

```python
df['month'].value_counts().plot(kind='bar')
plt.title('Highest Months For Sales')
plt.show()
```
